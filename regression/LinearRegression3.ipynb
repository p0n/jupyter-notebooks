```python
import numpy as np
from sklearn import datasets
from matplotlib import pyplot as plt
```

```python
# ボストンデータセットを読み込む
boston = datasets.load_boston()
# 部屋の数を取り出す
rooms = boston.data[:, 5]
# 家の値段を取り出す
house_prices = boston.target
```

```python
# 最小二乗法で誤差が最も少なくなる直線を得る
x = np.array([[v] for v in rooms])
y = house_prices
slope, total_error, _, _ = np.linalg.lstsq(x, y)
```

```python
# 訓練誤差の平均二乗平方根誤差 (Root Mean Squared Error: RMSE)
rmse = np.sqrt(total_error[0] / len(x))
msg = 'RMSE (training): {0}'.format(rmse)
print(msg)
```

```python
# 部屋の数と家の値段の関係をプロットする
plt.scatter(rooms, house_prices, color='r')

# 得られた直線をプロットする
plt.plot(x, slope * x)

# グラフを表示する
plt.xlabel('部屋の数')
plt.ylabel('家の値段 (単位: 1000 ドル)')
plt.autoscale()
plt.grid()
plt.show()
```
