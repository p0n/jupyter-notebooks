```python
import math
import bokeh.plotting as bplt
import numpy as np
```

```python
bplt.output_notebook()
```

```python
cycle = math.pi * 2
x = np.linspace(-1 * cycle, cycle, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
p = bplt.figure(title='sin/cos curve', plot_width=640, plot_height=320)
p.circle(x, y1, color='red')
p.triangle(x, y2, color='blue')
bplt.show(p)
```
